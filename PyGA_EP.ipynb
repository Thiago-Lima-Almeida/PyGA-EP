#Logística Empresarial
<h2>O aluno aprenderá sobre o tema na disciplina Logística, no 6º período, porém será apresentado no 1º período em Introdução à Engenharia de Produção.</h2>

A empresa também pediu que o engenheiro, por meio do plano cartesiano, representasse o fluxo dos processos, então com a disposição das máquinas mapeadas, restava a ordem das máquinas, que foi encontrada ao observar o procedimento de produção dos lanches.

Ordem das máquinas e seus posicionamentos em relação à origem do plano cartesiano:

*   **Máquina 1**: 3 metros à direita e 4 metros acima;
*   **Máquina 2**: 4 metros à esquerda e 5 metros acima;
*   **Máquina 3**: 1 metro à direita e 3 metros acima;
*   **Máquina 4**: 5 metros à direita e 2 metros abaixo;
*   **Máquina 5**: 2 metros à esquerda e 3 metros abaixo.

Vizualiando o gráfico, percebe-se que a movimentação está bastante agitada, o produto percorre caminhos contraintuitivos, por conseguinte, menos eficientes.

```python
import matplotlib.pyplot as plt

plt.figure()
plt.title("Fluxo de processos do fast-food")
plt.grid(zorder=0)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.xticks(range(-7, 8, 1))
plt.yticks(range(-7, 8, 1))

machinesX = [3,5,1,-4,-2]
machinesY = [4,-2,3,5,-3]
fluxo = [0,3,2,1,4]         # Ordem do fluxo, fluxo[0] representa a máquina 1.
rotulos = [f"M{i+1}" for i in range(len(fluxo))]        # Cria uma lista com o nome de cada máquina.

plt.scatter(machinesX, machinesY, s=150, c='green', edgecolors='black', linewidth=1.5, label = "Máquinas", zorder=2)

for fluxo_id, ponto_id in enumerate(fluxo):
    plt.text(machinesX[ponto_id] + 0.3, machinesY[ponto_id] + 0.8, rotulos[fluxo_id], fontsize=12, va='top')        # Nomeia cada máquina, colocando o nome acima do ponto.

for i in range(len(fluxo)-1):
    x1, y1 = machinesX[fluxo[i]], machinesY[fluxo[i]]         # Coordenadas da base da seta.
    x2, y2 = machinesX[fluxo[i + 1]], machinesY[fluxo[i + 1]]         # Coordenadas do fim da seta.
    plt.arrow(x1, y1, x2 - x1, y2 - y1, head_width=0.3, head_length=0.5, fc="black", ec="black", length_includes_head=True, zorder=3) # Cria uma seta partindo do ponto 1 para o ponto 2, do ponto 2 para o 3 e assim por diante.

plt.show()
```

#Modelagem, Simulação e Otimização
<h2>O tema será abordado no 5º período, durante as disciplinas Processos Estocásticos e Simulação de Sistemas. A resolução matemática apresentada será apresentada no 1º perído, em Geometria Analítica.</h2>

Para avaliar a eficiência do fluxo e mostrar um serviço diferenciado, além do que a empresa pediu, o engenheiro decidiu calcular as distâncias entre as máquinas na ordem de processamento, utilizando a fórmula de distância entre dois pontos: <br>
**$D_{ab} =\sqrt{(x_{b}-x_{a})^{2}+(y_{b}-y_{a})^{2}}$**

<h2>Resolução matemática</h2>
Utilizando a fórmula dada acima, tem-se:

* M1 para M2 <br>
$D_{M1M2}=\sqrt{(-4-3)^2+(5-4)^2}$<br>
$D_{M1M2}=\sqrt{(-7)^2+(1)^2}$<br>
$D_{M1M2}=\sqrt{49+1} = \sqrt{50} ≈ 7,07$ *u.d.*<br>
* M2 para M3 <br>
$D_{M2M3}=\sqrt{(1-(-4))^2+(3-5)^2}$ <br>
$D_{M2M3}=\sqrt{5^2+(-2)^2}$ <br>
$D_{M2M3}=\sqrt{25+4} = \sqrt{29} ≈ 5,39$ *u.d.* <br>
* M3 para M4 <br>
$D_{M3M4}=\sqrt{(5-1)^2+(-2-3)^2}$ <br>
$D_{M3M4}=\sqrt{4^2+(-5)^2}$ <br>
$D_{M3M4}=\sqrt{16+25} = \sqrt{41} ≈ 6,40$ *u.d.* <br>
* M4 para M5 <br>
$D_{M4M5}=\sqrt{(-2-5)^2+(-3-(-2))^2}$ <br>
$D_{M4M5}=\sqrt{(-7)^2+(-1)^2}$ <br>
$D_{M4M5}=\sqrt{49+1} = \sqrt{50} ≈ 7,07$ *u.d.* <br>

```python
import matplotlib.pyplot as plt
import math

plt.figure()
plt.title("Distâncias entre as máquinas")
plt.grid(zorder = 0)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.xticks(range(-7, 8, 1))
plt.yticks(range(-7, 8, 1))

machinesX = [3,5,1,-4,-2]
machinesY = [4,-2,3,5,-3]
fluxo = [0,3,2,1,4]
rotulos = ["M1","M2","M3","M4","M5"]

plt.scatter(machinesX, machinesY, s=150, c='green', edgecolors='black', linewidth=1.5, label = "Máquinas", zorder =2 )

for fluxo_id, ponto_id in enumerate(fluxo):
    plt.text(machinesX[ponto_id] + 0.3, machinesY[ponto_id] + 0.7, rotulos[fluxo_id], fontsize=12, va='top')

plt.plot([machinesX[i] for i in fluxo], [machinesY[i] for i in fluxo], linestyle=(0, (5, 10)), color = 'blue', label = "Fluxo", zorder = 2) # Cria uma linha tracejada entre as máquinas.

for i in range(len(fluxo) - 1):
    x1, y1 = machinesX[fluxo[i]], machinesY[fluxo[i]]
    x2, y2 = machinesX[fluxo[i + 1]], machinesY[fluxo[i + 1]]
    distancia = math.dist((x1, y1), (x2, y2))         # Utilização da função math.dist para calcular a distância euclidiana entre pontos.
    x_meio, y_meio = (x1 + x2) / 2, (y1 + y2) / 2         # Ponto central do fluxo entre 2 máquinas.
    plt.text(x_meio, y_meio, f"{distancia:.2f} u.d.", fontsize=12, color='black', ha='center', zorder =2)

plt.legend()
plt.show()
```

# Ergonomia
<h2>O tema será apresentado ao aluno no 1º período na disciplina de Introdução à Engenharia de Produção e aprofundado no 4º período, na disciplina de Ergonomia</h2>

O trabalhador realiza sua função numa máquina. A cada
20 acionamentos o trabalhador para a máquina e faz a

setup mais seguro, o Engenheiro de Segurança do Trabalho
estuda qual a melhor posição para posicionar uma botoeira
de emergência, responsável pela parada total da máquina.
Ao conversar com o trabalhador duas opções (A e B) foram
propostas (Figura 1). Qual das propostas apresenta menor
risco osteomuscular para a articulação do ombro? Sabe-se
que o centro de massa do braço (segmento entre o ombro e
cotovelo) concentra uma massa igual a 23 N; e que o
segmento do antebraço+mão (segmento entre o cotovelo e
dedos) concentra uma massa igual a 11 N.

<h2>Resolução Matemática</h2>
Utilizando o conceito de torque $(τ = \vec{F}*d)$ é possível determinar qual posição tem o menor risco osteomuscular, sendo $\vec{F}$ a força exercida e $d$ a distância do ponto de rotação até o ponto onde é aplicada a força.<br>
OBS: Ambas as posições possuem 2 momentos, sendo o momento 1 o segmento ombro-cotovelo e o momento 2 o segmento cotovelo-mão.

**Análise da Posição A**
* $\vec{τ}_{1} = 23*0,15 = 23*0,15 = 3,45N$
* $\vec{τ}_{2} = 11*0,30 = 11*0,30 = 3,30N$
* O torque total exercido na posição A se dá por $\vec{τ}_{t} = \vec{τ}_{1}+\vec{τ}_{2}$, logo, tem-se que $\vec{τ}_{t}=3,45+3,30 = 6,75N$

**Análise da Posição B**
* $\vec{τ}_{1} = 23*0,15 = 23*0,15 = 3,45N$ <br>
* $\vec{τ}_{2} = 11*0,15 = 11*0,15 = 1,65N$
* o torque total exercido na posição B será $\vec{τ}_{t}=3,45+1,65 = 5,10N$.

Logo, a posição B é a que apresenta um menor risco osteomuscular.

<h2>Análise dos ângulos da posição A em um Diagrama de Corpo Livre (DCL) de modo computacional</h2>


```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches        # Módulo utilizado para criar arcos

fig, (img1, img2) = plt.subplots(1, 2)        # fig é a janela principal e img 1 e 2 são seus subplots

fig.suptitle('Diagrama do Corpo Livre na Posição A.', fontsize=16, fontweight='bold')         # Título principal
img1.set_title('DCL do Ombro-Cotovelo (M1)')        # Título do subplot 1
img1.set_xticks([])         # Lista vazia para esconder a numeração do eixo x
img1.set_yticks([])         # Lista vazia para esconder a numeração do eixo y
img1.set_xlim(-10, 10)        # Limites do eixo x, de -10 à 10
img1.set_ylim(-10, 10)        # Limites do eixo y, de -10 à 10
img1.arrow(-7.5, 0, 15, 0, head_width=0.5, head_length=0.5, color='blue', label='Ombro-Cotovelo')         # Seta que representa a força exercida pelo segmento Ombro-Cotovelo
img1.arrow(0,0,0,-5, head_width=0.5, head_length=0.5, color='turquoise', label='Centro de Massa:23N')         # Seta que representa o centro de massa da região
img1.arrow(0,0,0,-3, head_width=0.5, head_length=0.5, color='black', linestyle='dotted', label='Força Gravitacional')         # Seta que representa a força da gravidade
arc1 = patches.Arc([0, 0], 2, 2, angle=0, theta1=270, theta2=0,color='red')         # Ângulo formado entre a força exercida e a força da gravidade
img1.add_patch(arc1)
img1.text(1, -1.5, s='90°')
img1.set_aspect('equal', adjustable='box')        # Regulando o subplot para manter uma proporção quadrada
img1.legend(loc='upper left')         # Adição de legenda no canto superior esquerdo

img2.set_title('DCL do Cotovelo-Mão (M2)')        # Título do subplot 2
img2.set_xticks([])         # Lista vazia para esconder a numeração do eixo x
img2.set_yticks([])         # Lista vazia para esconder a numeração do eixo y
img2.set_xlim(-10, 10)        # Limites do eixo x, de -10 à 10
img2.set_ylim(-10, 10)        # Limites do eixo y, de -10 à 10
img2.arrow(-10, 0, 2.5, 0, linestyle='dotted', color='blue')        # Seta pontilhada representando a seção Ombro-Cotovelo
img2.arrow(-7.5, 0, 15, 0, head_width=0.5, head_length=0.5, color='green', label='Cotovelo-Mão')         # Seta que representa a força exercida pelo segmento Cotovelo-Ombro
img2.arrow(0, 0, 0, -5, head_width=0.5, head_length=0.5, color='springgreen', label='Centro de Massa: 11N')          # Seta que representa o centro de massa da região
img2.arrow(0, 0, 0, -3, head_width=0.5, head_length=0.5, color='black', linestyle='dotted', label='Força Gravitacional')         # Seta que representa a força da gravidade
arc2 = patches.Arc([0, 0], 2, 2, angle=0, theta1=270, theta2=0, color='red')         # Ângulo formado entre a força exercida e a força da gravidade
img2.text(1, -1.5, s='90°')
img2.add_patch(arc2)
img2.set_aspect('equal', adjustable='box')        # Regulando o subplot para manter uma proporção quadrada
img2.legend(loc='upper left')         # Adição de legenda no canto superior esquerdo

plt.tight_layout()

plt.show()
```

* Momento 1:
<br>A força está sendo exercida horizontalmente para a direita e a força da gravidade aponta sempre para baixo, logo, o ângulo $θ_{OcFg}$ formado entre os 2 vetores é $90°$.
* Momento 2:
<br>A força está sendo aplicada novamente na horizontal para a direita, então o ângulo $θ_{CmFg}$ também será $90°$.
