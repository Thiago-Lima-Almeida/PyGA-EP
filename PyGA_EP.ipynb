# Projeto de Fábrica e Instalações Industriais

<h3>Representação de um layout industrial no plano cartesiano</h3>

Um Engenheiro de Produção foi contratado por uma rede de Fast-Food para representar o layout de uma unidade da rede, por meio de um plano cartesiano, com o intuito de identificar espaços vagos já que não é possível verificar pessoalmente, uma vez que isso atrapalharia o funcionamento do estabelecimento.

Ao mapear o local, ele descobriu o centro do fast-food, a posição de cada máquina e ao representar no plano cartesiano, obteve o seguinte gráfico:


```python
import matplotlib.pyplot as plt # Pyplot é um módulo que gera gráficos interativos.

plt.figure()        # Cria uma figura
plt.title("Chão de fábrica da unidade X do Fast-Food Y")        # Define o título do gráfico.
plt.grid(zorder = 0)        # plt.grid() ativa as linhas de grade, zorder = 0 define a camada do elemento na figura, sendo 0 a menor.
plt.axhline(0, color='black')         # Cria uma linha horizontal no eixo y, na posição 0.
plt.axvline(0, color='black')         # Cria uma linha vertical no eixo x, na posição 0.
plt.xlim(-6, 6)         # Delimita o início e fim do eixo x
plt.ylim(-6, 6)         # Delimita o início e fim do eixo y
plt.xticks(range(-6, 7, 1))         # Define os valores que serão exibidos no eixo x, em intervalos de 1 em 1.
plt.yticks(range(-6, 7, 1))         # Define os valores que serão exibidos no eixo y, em intervalos de 1 em 1.

machinesX = [3,5,1,-4,-2]         # Posições das máquinas no eixo x.
machinesY = [4,-2,3,5,-3]         # Posições das máquinas no eixo y.

plt.scatter(machinesX, machinesY, s=450, c='green', marker='s', edgecolors='black', linewidth=1.5, zorder = 2)        # Cria um gráfico de dispersão com a coordenada das máquinas
plt.show()        # Comando para exibir o gráfico corretamente.
```

#Logística Empresarial

<h3>Representação de fluxos por meio de vetores</h3>


A empresa também pediu que o engenheiro, por meio do plano cartesiano, representasse o fluxo dos processos, então com a disposição das máquinas mapeadas, restava a ordem das máquinas, que foi encontrada ao observar o procedimento de produção dos lanches.

Ordem das máquinas e seus posicionamentos em relação à origem do plano cartesiano:

*   **Máquina 1**: 3 metros à direita e 4 metros acima;
*   **Máquina 2**: 4 metros à esquerda e 5 metros acima;
*   **Máquina 3**: 1 metro à direita e 3 metros acima;
*   **Máquina 4**: 5 metros à direita e 2 metros abaixo;
*   **Máquina 5**: 2 metros à esquerda e 3 metros abaixo.

Vizualiando o gráfico, percebe-se que a movimentação está bastante agitada, o produto percorre caminhos contraintuitivos, por conseguinte, menos eficientes.

```python
import matplotlib.pyplot as plt

plt.figure()
plt.title("Fluxo de processos do fast-food")
plt.grid(zorder=0)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.xticks(range(-7, 8, 1))
plt.yticks(range(-7, 8, 1))

machinesX = [3,5,1,-4,-2]
machinesY = [4,-2,3,5,-3]
fluxo = [0,3,2,1,4] # Ordem do fluxo, fluxo[0] representa a máquina 1.
rotulos = [f"M{i+1}" for i in range(len(fluxo))] # Cria uma lista com o nome de cada máquina.

plt.scatter(machinesX, machinesY, s=150, c='green', edgecolors='black', linewidth=1.5, label = "Máquinas", zorder=2)

for fluxo_id, ponto_id in enumerate(fluxo):
    plt.text(machinesX[ponto_id] + 0.3, machinesY[ponto_id] + 0.8, rotulos[fluxo_id], fontsize=12, va='top') # Nomeia cada máquina, colocando o nome acima do ponto.

for i in range(len(fluxo)-1):
    x1, y1 = machinesX[fluxo[i]], machinesY[fluxo[i]]         # Coordenadas da base da seta.
    x2, y2 = machinesX[fluxo[i + 1]], machinesY[fluxo[i + 1]]         # Coordenadas do fim da seta.
    plt.arrow(x1, y1, x2 - x1, y2 - y1, head_width=0.3, head_length=0.5, fc="black", ec="black", length_includes_head=True, zorder=3) # Cria uma seta partindo do ponto 1 para o ponto 2, do ponto 2 para o 3 e assim por diante.

plt.show()
```

#Modelagem, Simulação e Otimização

<h3>Cálculo de distâncias</h3>

Para avaliar a eficiência do fluxo e mostrar um serviço diferenciado, além do que a empresa pediu, o engenheiro decidiu calcular as distâncias entre as máquinas na ordem de processamento, utilizando a fórmula de distância entre dois pontos: <br>
**$D_{ab} =\sqrt{(x_{b}-x_{a})^{2}+(y_{b}-y_{a})^{2}}$**

<h2>Resolução matemática</h2>
Utilizando a fórmula dada acima, tem-se:

* M1 para M2 <br>
$D_{M1M2}=\sqrt{(-4-3)^2+(5-4)^2}$<br>
$D_{M1M2}=\sqrt{(-7)^2+(1)^2}$<br>
$D_{M1M2}=\sqrt{49+1} = \sqrt{50} ≈ 7,07$ *u.d.*<br>
* M2 para M3 <br>
$D_{M2M3}=\sqrt{(1-(-4))^2+(3-5)^2}$ <br>
$D_{M2M3}=\sqrt{5^2+(-2)^2}$ <br>
$D_{M2M3}=\sqrt{25+4} = \sqrt{29} ≈ 5,39$ *u.d.* <br>
* M3 para M4 <br>
$D_{M3M4}=\sqrt{(5-1)^2+(-2-3)^2}$ <br>
$D_{M3M4}=\sqrt{4^2+(-5)^2}$ <br>
$D_{M3M4}=\sqrt{16+25} = \sqrt{41} ≈ 6,40$ *u.d.* <br>
* M4 para M5 <br>
$D_{M4M5}=\sqrt{(-2-5)^2+(-3-(-2))^2}$ <br>
$D_{M4M5}=\sqrt{(-7)^2+(-1)^2}$ <br>
$D_{M4M5}=\sqrt{49+1} = \sqrt{50} ≈ 7,07$ *u.d.* <br>

```python
import matplotlib.pyplot as plt
import math

plt.figure()
plt.title("Distâncias entre as máquinas")
plt.grid(zorder = 0)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.xticks(range(-7, 8, 1))
plt.yticks(range(-7, 8, 1))

machinesX = [3,5,1,-4,-2]
machinesY = [4,-2,3,5,-3]
fluxo = [0,3,2,1,4]
rotulos = ["M1","M2","M3","M4","M5"]

plt.scatter(machinesX, machinesY, s=150, c='green', edgecolors='black', linewidth=1.5, label = "Máquinas", zorder =2 )

for fluxo_id, ponto_id in enumerate(fluxo):
    plt.text(machinesX[ponto_id] + 0.3, machinesY[ponto_id] + 0.7, rotulos[fluxo_id], fontsize=12, va='top')

plt.plot([machinesX[i] for i in fluxo], [machinesY[i] for i in fluxo], linestyle=(0, (5, 10)), color = 'blue', label = "Fluxo", zorder = 2) # Cria uma linha tracejada entre as máquinas.

for i in range(len(fluxo) - 1):
    x1, y1 = machinesX[fluxo[i]], machinesY[fluxo[i]]
    x2, y2 = machinesX[fluxo[i + 1]], machinesY[fluxo[i + 1]]
    distancia = math.dist((x1, y1), (x2, y2))         # Utilização da função math.dist para calcular a distância euclidiana entre pontos.
    x_meio, y_meio = (x1 + x2) / 2, (y1 + y2) / 2         # Ponto central do fluxo entre 2 máquinas.
    plt.text(x_meio, y_meio, f"{distancia:.2f} u.d.", fontsize=12, color='black', ha='center', zorder =2)

plt.legend()
plt.show()
```

<h3>Cálculo de ângulos para aferir a postura</h3>

O trabalhador realiza sua função numa máquina. A cada
20 acionamentos o trabalhador para a máquina e faz a
limpeza nos pontos de operação. Na tentativa tornar tal
setup mais seguro, o Engenheiro de Segurança do Trabalho
estuda qual a melhor posição para posicionar uma botoeira
de emergência, responsável pela parada total da máquina.
Ao conversar com o trabalhador duas opções (A e B) foram
propostas (Figura 1). Qual das propostas apresenta menor
risco osteomuscular para a articulação do ombro? Sabe-se
que o centro de massa do braço (segmento entre o ombro e
cotovelo) concentra uma massa igual a 23 N; e que o
segmento do antebraço+mão (segmento entre o cotovelo e
dedos) concentra uma massa igual a 11 N.

<h2>Resolução Matemática</h2>
Utilizando o conceito de torque $(τ = F*d*sen(θ_{FgF}))$ é possível determinar qual posição tem o menor risco osteomuscular, sendo F a força exercida, d a distância do ponto de rotação até o ponto onde é aplicada a força e o seno do ângulo entre a força aplicada e a força gravitacional (sempre para baixo). <br>
OBS: Ambas as posições possuem 2 momentos, sendo o momento 1 o segmento ombro-cotovelo e o momento 2 o segmento cotovelo-mão.

**Análise da Posição A**
* Na posição A, as forças estão sendo aplicadas na horizontal, logo o ângulo entre a força aplicada e a força gravitacional (que age verticalmente) é de 90º, então:<br>
$τ_{1} = 23×0,15×sen(90º) = 23×0,15×1 = 3,45N$
$τ_{2} = 11×0,30×sen(90º) = 11×0,30×1 = 3,30N$
* O torque total exercido na posição A se dá por $τ_{t} = τ_{1}+τ_{2}$, logo, tem-se que $τ_{t}=3,45+3,30 = 6,75N$

**Análise da Posição B**
* Na posição B, as forças estão sendo aplicadas em eixos diferentes, o segmento ombro-cotovelo aplica uma força horizontal enquanto o segmento cotovelo-mão aplica uma força vertical, portanto no momento 1, o ângulo formado com a força gravitacional é de 90º, enquanto no momento 2, o ângulo formado com a força gravitacional é 0º visto que a força aplicada também age verticalmente, logo são paralelas. <br>
$τ_{1} = 23×0,15×sen(90º) = 23×0,15×1 = 3,45N$
$τ_{2} = 11×0,30×sen(0) = 11×0,30×0 = 0N$
* o torque total exercido na posição B será $τ_{t}=3,45+0 = 3,45N$.

Logo, a posição B é a que apresenta um menor risco osteomuscular.
